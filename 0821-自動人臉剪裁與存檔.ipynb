```python
casc_path = cv2.data.haarcascades+"haarcascade_frontalface_default.xml"
detector = cv2.CascadeClassifier(casc_path)           # 載入人臉追蹤模型
recog = cv2.face.LBPHFaceRecognizer_create()          # 啟用訓練人臉模型方法
faces = []                                            # 儲存人臉位置大小的串列
ids = []                                              # 記錄該人臉 id 的串列
```

```python
for s, i in enumerate(imgs):
    im = cv2.imread(i)
    gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    img_np = np.array(gray,'uint8')
    #print(gray)
    #print(img_np)
    face = detector.detectMultiScale(img_np,
                                    scaleFactor=1.1,
                                    minNeighbors=5,
                                    minSize=(30,30),
                                    flags = cv2.CASCADE_SCALE_IMAGE)
    
    #print(s)
    #print(i)
    #print(face)
    #print(len(face))
    #print(face[0])
    

    
    for(x,y,w,h) in face:
        faces.append(img_np[y:y+h,x:x+w])
        cv2.imwrite("C:/Users/Wayne/Desktop/media\\face"+ str(s) +".jpg" , faces[s])
        
#print(faces[0])
#cv2.imshow("test",faces[5])
#cv2.waitKey (0)  
#cv2.imwrite("C:/Users/Wayne/Desktop/media\\face.jpg",faces[0])
```
